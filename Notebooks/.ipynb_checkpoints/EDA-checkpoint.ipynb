# Dataset Description

```python
import os 

dir = '../Dataset/flowers/'
total_data = 0

def filter_jpg_files(files):
    """Filters a list of files for files with a .jpg format.

    Args:
        files: A list of files.

    Returns:
        A list of files with a .jpg format.
    """

    jpg_files = []
    for file in files:
        if file.endswith(".jpg"):
            jpg_files.append(file)
    return jpg_files


elements = dict()
print("*"*30)
print("Data Instances:")
for classes in os.listdir(dir):
    len_data = len(filter_jpg_files(os.listdir(dir+classes)))
    total_data+=len_data
    elements[classes] = len_data
    print(classes,":",len_data)
print("*"*30)
print("Total Data:",total_data)
print("*"*30)
print("Total Classes:",len(elements))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dictionary of flower counts
flower_counts = elements

# Create a list of the keys and values in the dictionary
keys = list(flower_counts.keys())
values = list(flower_counts.values())

# Create a barplot using Seaborn's barplot() function
sns.barplot(x=keys, y=values)

# Add a title and axis labels to the plot
plt.title("Flower Counts")
plt.xlabel("Flower Type")
plt.ylabel("Count")

# Show the plot
plt.show()
```

## Relationship between different classes and average pixel values

```python
import cv2
import glob
import numpy as np

def calc_stat(img_data):
    mean_red = np.mean(img_data[:, :, 0])
    std_red = np.std(img_data[:, :, 0])
    median_red = np.median(img_data[:, :, 0])
    
    mean_green = np.mean(img_data[:, :, 1])
    std_green = np.std(img_data[:, :, 1])
    median_green = np.median(img_data[:, :, 1])
    
    mean_blue = np.mean(img_data[:, :, 2])
    std_blue = np.std(img_data[:, :, 2])
    median_blue = np.median(img_data[:, :, 2])

    # Print the statistics
    print("Mean red:", mean_red)
    print("Standard deviation red:", std_red)
    print("Median red:", median_red)
    
    print("Mean green:", mean_green)
    print("Standard deviation green:", std_green)
    print("Median green:", median_green)
    
    print("Mean blue:", mean_blue)
    print("Standard deviation blue:", std_blue)
    print("Median blue:", median_blue)

def display_rel(path):
    img_data = []
    image_paths = glob.glob(path)
    image_size = (224, 224)
    resized_images = []

    for image_path in image_paths:
        image = cv2.imread(image_path)
        image = cv2.resize(image, image_size)
        resized_images.append(image)

    img_data = np.array(resized_images)

    hist_red = np.histogram(img_data[:, :, 0], bins=256)
    hist_green = np.histogram(img_data[:, :, 1], bins=256)
    hist_blue = np.histogram(img_data[:, :, 2], bins=256)

    plt.plot(hist_red[1][:-1], hist_red[0])
    plt.plot(hist_green[1][:-1], hist_green[0])
    plt.plot(hist_blue[1][:-1], hist_blue[0])
    
    # Add labels and title
    plt.xlabel("Pixel value")
    plt.ylabel("Frequency")
    plt.title("Histogram of pixel values")
    
    # Show the plot
    plt.show()
    calc_stat(img_data)
```

```python
for classes in os.listdir(dir):
    print("*"*45)
    print(classes)
    display_rel(dir+classes+"/*.jpg")
```

## Dataset Preprocessing
* Since the dataset is imbalanced we need to balance the dataset.   
The techniques used here is the combination of undersampling and oversampling. 

* First of all the average is found out by which the undersampling and oversampling can be splitted

```python
avg = sum(elements.values())//5
print("Average",avg)
```

```python
import tensorflow as tf
def oversampling(path, no_img_aug):
    
    image_data_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
    )
    
    augmented_images = image_data_generator.flow_from_directory(
    directory=path,
    batch_size=no_img_aug,
    target_size=(224, 224)
    )

    # Save the augmented images to a new directory.
    augmented_images_directory = path
    
    if not os.path.exists(augmented_images_directory):
        os.makedirs(augmented_images_directory)

    for i, image in enumerate(augmented_images):
        image_path = os.path.join(augmented_images_directory, '{}.jpg'.format(i))
        
        tf.keras.preprocessing.image.save_img(image_path, image)
```

```python

def undersampling(path, no_img_aug, no_img_rm):
    files = os.listdir(path)
    for cnt,f in enumerate(files):
        if cnt>no_img_rm:
            break
        os.remove(f)
```

```python
import tensorflow as tf

# Create an instance of the ImageDataGenerator class.
image_data_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Generate a batch of augmented images.
augmented_images = image_data_generator.flow_from_directory(
    directory='/path/to/images/',
    batch_size=32,
    target_size=(224, 224)
)

# Save the augmented images to a new directory.
augmented_images_directory = '/path/to/augmented_images/'
if not os.path.exists(augmented_images_directory):
    os.makedirs(augmented_images_directory)

for i, image in enumerate(augmented_images):
    image_path = os.path.join(augmented_images_directory, '{}.jpg'.format(i))
    tf.keras.preprocessing.image.save_img(image_path, image)
```
